 Load audio files and extract features

```python
def parser(row):
    file_name = os.path.join(os.path.abspath('H:\\Datatsets\\Audio Processing\\Urban Sound Classification\\'), 
                             'Train', str(row.ID) + '.wav') # function to load files and extract features
    try:  # handle exception to check if there isn't a file which is corrupted
        X, sample_rate = librosa.load(file_name, res_type='kaiser_fast')  # here kaiser_fast is a technique used for faster extraction
        mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0) # we extract mfcc feature from data
    except Exception as e:
        print("Error encountered while parsing file: ", file_name)
        return None, None
    
    feature = mfccs
    label = row.Class
    
    return pd.Series([feature, label])

temp = train.apply(parser, axis=1)
temp.columns = ['feature', 'label']
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
from keras import utils as np_utils

X = np.array(temp.feature.tolist())
y = np.array(temp.label.tolist())

lb = LabelEncoder()

y = np_utils.to_categorical(lb.fit_transform(y))
```

```python
# build model
model = Sequential()

model.add(Dense(256, input_shape=X.shape[1:]))
model.add(Activation('relu'))
model.add(Dropout(0.2))

model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dropout(0.2))

model.add(Dense(num_labels))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```
